# **LLM integration**

# Single Text Translation

```python
import google.generativeai as genai

genai.configure(api_key="your_api_key")
model = genai.GenerativeModel("gemini-1.5-flash")
```

```python
def translate_single_text(json_input):
    text = json_input.get("text", "") # Lấy text của người dùng dưới dạng json
    dest_language = json_input.get("dest_language", "vi")

    if not text or not dest_language:
        raise ValueError("JSON input phải chứa 'text' và 'dest_language'.")

    # Nếu văn bản đã là tiếng Việt thì trả về chính nó
    if dest_language == "vi" and text.isascii() is False:
        return text

    # Gửi yêu cầu tới mô hình
    prompt = f" Only Translate this text to {dest_language}: {text}" # Yêu cầu chỉ dịch text sang tiếng Việt thôi
    response = model.generate_content(prompt)
    return response.text.strip() #Lấy nội dung phản hồi từ API và loại bỏ khoảng trắng dư thừa hoặc ký tự không cần thiết (strip)
```

```python
json_1 = {
'text': 'Hello',
'dest_language': 'vi'
}

translate_single_text(json_1)
```

# Multiple Texts Translation

```python
def translate_multiple_texts(json_input):
    texts = json_input.get("text", []) #Lấy text của người dùng dưới dạng json
    dest_language = json_input.get("dest_language", "vi")

    if not texts or not dest_language:
        raise ValueError("JSON input phải chứa 'text' (danh sách) và 'dest_language'.")

    if not isinstance(texts, list):
        raise ValueError("'text' trong JSON input phải là một danh sách.")

    results = [] #tạo danh sách rỗng
    for text in texts:
        # Nếu văn bản đã là tiếng Việt thì thêm nó dô danh sách
        if dest_language == "vi" and text.isascii() is False:
            results.append(text)
        else:
            # Gửi yêu cầu tới mô hình
            prompt = f"Only Translate this text to {dest_language}: {text}"  # Yêu cầu chỉ dịch text sang tiếng Việt thôi
            response = model.generate_content(prompt)
            results.append(response.text.strip()) #thêm text được dịch vào danh sáchh
    return results #trả về kết quả
```

```python
json_2 = {
'text': ['Hello','các bạn', 'I am Peter'],
'dest_language': 'vi'
}

translate_multiple_texts(json_2)
```
